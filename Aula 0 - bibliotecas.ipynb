## O que são bibliotecas?

Em Python, uma **biblioteca** (também chamada de *pacote* ou *módulo*) é um conjunto de código pronto que outras pessoas já escreveram e disponibilizaram para reutilização. Em vez de implementar tudo do zero, você **importa** funções, classes e ferramentas que já existem.

### Como importar uma biblioteca

```python
import pandas                                   #importa toda a biblioteca 
import numpy as np                              # importamoso numpy com o apelido de "np"
from matplotlib import pyplot as plt               #importa só o módulo pyplot
from sklearn.linear_model import LinearRegression  # importa só uma função ou classe específica
``

## Como instalar bibliotecas?

A biblioteca padrão do Python já vem com módulos básicos (`math`, `os`, `datetime`...). Para pacotes externos, usamos gerenciadores de pacotes.

### 1. `pip` (mais comum)

O **pip** é o instalador oficial de pacotes Python. Use no terminal (Prompt de Comando, PowerShell ou terminal do VS Code/Cursor):

```bash
pip install numpy
pip install pandas matplotlib scikit-learn
pip install nome-do-pacote==1.2.3   # versão específica
pip install -r requirements.txt   # lista de pacotes de um arquivo
pip list                          # ver pacotes instalados
pip show pandas                   # detalhes de um pacote
pip uninstall numpy               # remover um pacote
```

## NumPy

O [Numpy](https://numpy.org) fornece o tipo **`ndarray`** (array multidimensional) e operações vetorizadas.

````python
import numpy as np

# Criando arrays
arr = np.array([1, 2, 3, 4, 5])
matriz = np.array([[1, 2, 3],[4, 5, 6]])

print("Array:", arr)
print("Matriz:\n", matriz)
print("Forma:", matriz.shape)
print("Média:", arr.mean())
print("Soma:", arr.sum())

print("\nArray × 2:", arr * 2)
print("Raiz quadrada:", np.sqrt(arr))
````

## Pandas

O [pandas](https://pandas.pydata.org) é a biblioteca mais usada para limpar, filtrar, agrupar e analisar dados antes de modelar ou visualizar.

````python
import pandas as pd

# Criando um DataFrame de exemplo
dados = {
    "nome": ["Ana", "Bruno", "Carla", "Diego"],
    "idade": [25, 30, 28, 35],
    "salario": [4500, 5200, 4800, 6100]
}
df = pd.DataFrame(dados)

print("DataFrame completo:")
display(df)

print("\nPrimeiras linhas (head):")
display(df.head(2))

print("\nEstatísticas descritivas:")
display(df.describe())

print("\nFiltrar idade > 28:")
display(df[df["idade"] > 28])

print("\nMédia salarial por faixa etária (> 30 anos):")
display(df[df["idade"] > 30]["salario"].mean())
````

## Matplotlib

# ![Exemplo de plot gerado com Matplotlib](https://coding-blocks.github.io/DS-NOTES/_images/matplotlib1.png)


O [matplotlib.org](https://matplotlib.org) é a biblioteca clássica de visualização em Python e é usada para gerar visualizações dos dados.

> Em projetos mais modernos, muitas pessoas também usam **Seaborn** (construído sobre o Matplotlib), mas o Matplotlib continua sendo fundamental.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
x = np.linspace(0, 10, 50)
y = np.sin(x)

# Gráfico de linha
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="sen(x)", color="steelblue", linewidth=2)
plt.title("Gráfico de linha — função seno")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Gráfico de dispersão
np.random.seed(42)
x_disp = np.random.randn(50)
y_disp = 2 * x_disp + np.random.randn(50) * 0.5

plt.figure(figsize=(8, 4))
plt.scatter(x_disp, y_disp, alpha=0.7, color="coral")
plt.title("Gráfico de dispersão")
plt.xlabel("Variável X")
plt.ylabel("Variável Y")
plt.show()
````

## Scikit-learn

O  [scikit-learn](https://scikit-learn.org) ( ou *sklearn*) reúne algoritmos de aprendizado de máquina e funções e algoritmos para **Classificação** (prever categorias), **Regressão** (prever valores numéricos), **Clustering** (agrupar dados sem rótulo), **Redução de dimensionalidade** (PCA, etc.) e **Pré-processamento** (padronização, divisão treino/teste)...

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Carrega dataset íris
iris = load_iris()
X, y = iris.data, iris.target  # X = features, y = rótulos

# Divide em treino (75%) e teste (25%)
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.25)

# Treina um modelo de classificação KNN
modelo = KNeighborsClassifier(n_neighbors=2)
modelo.fit(X_treino, y_treino)

# Faz previsões e avalia
y_pred = modelo.predict(X_teste)
print(f"Acurácia no teste: {accuracy_score(y_teste, y_pred):.2%}")
print("\nRelatório de classificação:")
print(classification_report(y_teste, y_pred, target_names=iris.target_names))
````
